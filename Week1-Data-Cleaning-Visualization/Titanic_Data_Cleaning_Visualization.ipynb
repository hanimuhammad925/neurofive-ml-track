## Clean & Visualize Real-World Data (Titanic Dataset)

**Objective:** Clean missing data, detect outliers, visualize patterns, and understand the dataset before machine learning.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Import Libraries

We use pandas and NumPy for data handling, while matplotlib and seaborn are used for data visualization.

```python
df = pd.read_csv("train.csv")
df.head()
```

```python
df.isnull().sum()
```

### Missing Values

The dataset contains missing values in Age, Cabin, and Embarked.

```python
df['Age']=df['Age'].fillna(df['Age'].median())
df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode()[0])
df.drop(columns=['Cabin'])
```

### Why?

Age was filled with the median because it is a numerical feature.

Embarked was filled with the mode because it is categorical.

Cabin was removed because most of its values were missing.

```python
df.isnull().sum()
```

# Histogram

```python
plt.figure(figsize=(10,6))
plt.hist(df["Age"], bins =20, color="blue", edgecolor="black")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

### Histogram Observation

This chart shows the distribution of passenger ages.

# Boxplot

```python
plt.figure(figsize=(8,3))

sns.boxplot(x=df["Fare"])

plt.title("Boxplot of Fare")
plt.xlabel("Fare")

plt.show()
```

### Boxplot Observation 

The Fare column contains several outliers. Most passengers paid lower fares, while a few passengers paid very high fares. These high values appear as separate points in the boxplot.

# Bar Chart


```python
plt.figure(figsize=(6,4))

sns.countplot(x="Survived", data=df)

plt.title("Survival Count")

plt.show()
```

### Bar Chart Observation

The chart shows that the number of passengers who did not survive is higher than the number of passengers who survived.

# Correlation Heatmap

```python
plt.figure(figsize=(8,6))

sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="Blues")

plt.title("Correlation Heatmap")

plt.show()
```

### Heatmap Observation

The heatmap shows relationships between numerical features. It indicates that Fare has a positive correlation with Survival, while Pclass has a negative correlation. Most other features have weak correlations.
