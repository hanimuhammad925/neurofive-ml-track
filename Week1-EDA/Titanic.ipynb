## Exploratory Data Analysis (EDA)

EDA (Exploratory Data Analysis) is the process of understanding a dataset before applying machine learning. It helps us identify the data structure, missing values, data types, and basic statistics.


## Import Libraries

Import the required libraries for data analysis.

```python
import pandas as pd
import numpy as np 
```

## Load Dataset

Load the Titanic dataset into a pandas DataFrame.

```python
df = pd.read_csv("train.csv")
```

## Preview Dataset

Display the first five rows of the dataset.

```python
df.head()
```

## Dataset Information

View the dataset structure, data types, and missing values.

```python
df.info()
```

## Statistical Summary

Generate summary statistics for numerical columns.

```python
df.describe()
```

## Dataset Shape

Check the number of rows and columns.

```python
df.shape
```

## Column Names

Display all column names.

```python
df.columns
```

## Data Types

Identify numerical and categorical columns.

```python
df.dtypes
```

```python
numerical = df.select_dtypes(include=['int64', 'float64']).columns
categorical = df.select_dtypes(include=['object']).columns
print("Numerical columns:", numerical)
print("\nCategorical columns:", categorical)
```

## Missing Values

Count missing values in each column.

```python
df.isnull().sum()
```

## Data Story

The Titanic dataset contains passenger information and their survival status. It has 891 rows and 12 columns. Some columns, such as Age and Cabin, contain missing values. The dataset includes both numerical and categorical features. This analysis provides a basic understanding of the data before preprocessing and machine learning.
